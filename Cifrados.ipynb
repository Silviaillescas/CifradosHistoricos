```python
def cesar_cifrar(mensaje, desplazamiento):
    resultado = ""
    
    for caracter in mensaje:
        
        if 'A' <= caracter <= 'Z':
            indice = ord(caracter) - ord('A')
            nuevo = (indice + desplazamiento) % 26
            resultado += chr(ord('A') + nuevo)
        
        elif 'a' <= caracter <= 'z':
            indice = ord(caracter) - ord('a')
            nuevo = (indice + desplazamiento) % 26
            resultado += chr(ord('a') + nuevo)
        
        else:
            resultado += caracter
    
    return resultado


def cesar_descifrar(mensaje, desplazamiento):
    return cesar_cifrar(mensaje, -desplazamiento)


# Prueba
mensaje = "QUE HACE"
cifrado = cesar_cifrar(mensaje, 3)
descifrado = cesar_descifrar(cifrado, 3)

print("Original:", mensaje)
print("Cifrado:", cifrado)
print("Descifrado:", descifrado)

```

```python
def rot13(mensaje):
    resultado = ""
    
    for caracter in mensaje:
        
        if 'A' <= caracter <= 'Z':
            indice = ord(caracter) - ord('A')
            nuevo = (indice + 13) % 26
            resultado += chr(ord('A') + nuevo)
        
        elif 'a' <= caracter <= 'z':
            indice = ord(caracter) - ord('a')
            nuevo = (indice + 13) % 26
            resultado += chr(ord('a') + nuevo)
        
        else:
            resultado += caracter
    
    return resultado


# Prueba
mensaje = "HELLO"
cifrado = rot13(mensaje)
doble = rot13(cifrado)

print("Original:", mensaje)
print("ROT13:", cifrado)
print("Aplicando ROT13 dos veces:", doble)
```

```python
def generar_desplazamientos(clave):
    desplazamientos = []
    
    for letra in clave:
        if letra.isalpha():
            valor = ord(letra.upper()) - ord('A')
            desplazamientos.append(valor)
    
    return desplazamientos


def vigenere_cifrar(mensaje, clave):
    resultado = ""
    desplazamientos = generar_desplazamientos(clave)
    
    j = 0
    
    for caracter in mensaje:
        
        if caracter.isalpha():
            k = desplazamientos[j % len(desplazamientos)]
            
            if 'A' <= caracter <= 'Z':
                indice = ord(caracter) - ord('A')
                nuevo = (indice + k) % 26
                resultado += chr(ord('A') + nuevo)
            
            elif 'a' <= caracter <= 'z':
                indice = ord(caracter) - ord('a')
                nuevo = (indice + k) % 26
                resultado += chr(ord('a') + nuevo)
            
            j += 1
        
        else:
            resultado += caracter
    
    return resultado


def vigenere_descifrar(mensaje, clave):
    resultado = ""
    desplazamientos = generar_desplazamientos(clave)
    
    j = 0
    
    for caracter in mensaje:
        
        if caracter.isalpha():
            k = desplazamientos[j % len(desplazamientos)]
            
            if 'A' <= caracter <= 'Z':
                indice = ord(caracter) - ord('A')
                nuevo = (indice - k) % 26
                resultado += chr(ord('A') + nuevo)
            
            elif 'a' <= caracter <= 'z':
                indice = ord(caracter) - ord('a')
                nuevo = (indice - k) % 26
                resultado += chr(ord('a') + nuevo)
            
            j += 1
        
        else:
            resultado += caracter
    
    return resultado


# Prueba 
mensaje = "MANIA"
clave = "LEMON"

cifrado = vigenere_cifrar(mensaje, clave)
descifrado = vigenere_descifrar(cifrado, clave)

print("Original:", mensaje)
print("Cifrado:", cifrado)
print("Descifrado:", descifrado)
```

```python
def analisis_frecuencia(mensaje):
    conteo = {}
    total = 0
    
    for caracter in mensaje:
        if caracter.isalpha():
            letra = caracter.upper()
            
            if letra in conteo:
                conteo[letra] += 1
            else:
                conteo[letra] = 1
            
            total += 1
    
    print("Total de letras:", total)
    print("\nTabla de frecuencia:")
    
    for letra in sorted(conteo):
        porcentaje = (conteo[letra] / total) * 100
        print(letra, ":", conteo[letra], "-", round(porcentaje, 2), "%")
    
    return conteo


# Prueba
analisis_frecuencia("AAABBC")
```
