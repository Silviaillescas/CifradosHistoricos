```python
import string
from collections import Counter

ABC_UPPER = string.ascii_uppercase
ABC_LOWER = string.ascii_lowercase

def _mod(n, m):
    return n % m

def _shift_char(c, k):
    # desplaza solo letras (A-Z / a-z), preserva mayúscula/minúscula
    if c in ABC_UPPER:
        i = ord(c) - ord('A')
        return chr(ord('A') + _mod(i + k, 26))
    if c in ABC_LOWER:
        i = ord(c) - ord('a')
        return chr(ord('a') + _mod(i + k, 26))
    return c

def _solo_letras(clave):
    return "".join(ch for ch in clave if ch.isalpha())
```

```python
def cesar_cifrar(mensaje, desplazamiento):
    resultado = ""
    
    for caracter in mensaje:
        
        if 'A' <= caracter <= 'Z':
            indice = ord(caracter) - ord('A')
            nuevo = (indice + desplazamiento) % 26
            resultado += chr(ord('A') + nuevo)
        
        elif 'a' <= caracter <= 'z':
            indice = ord(caracter) - ord('a')
            nuevo = (indice + desplazamiento) % 26
            resultado += chr(ord('a') + nuevo)
        
        else:
            resultado += caracter
    
    return resultado


def cesar_descifrar(mensaje, desplazamiento):
    return cesar_cifrar(mensaje, -desplazamiento)


# Prueba
mensaje = "QUE HACE"
cifrado = cesar_cifrar(mensaje, 3)
descifrado = cesar_descifrar(cifrado, 3)

print("Original:", mensaje)
print("Cifrado:", cifrado)
print("Descifrado:", descifrado)

```

```python
def rot13(mensaje):
    resultado = ""
    
    for caracter in mensaje:
        
        if 'A' <= caracter <= 'Z':
            indice = ord(caracter) - ord('A')
            nuevo = (indice + 13) % 26
            resultado += chr(ord('A') + nuevo)
        
        elif 'a' <= caracter <= 'z':
            indice = ord(caracter) - ord('a')
            nuevo = (indice + 13) % 26
            resultado += chr(ord('a') + nuevo)
        
        else:
            resultado += caracter
    
    return resultado


# Prueba
mensaje = "HELLO"
cifrado = rot13(mensaje)
doble = rot13(cifrado)

print("Original:", mensaje)
print("ROT13:", cifrado)
print("Aplicando ROT13 dos veces:", doble)
```
